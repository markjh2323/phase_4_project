```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_pickle('processed_df.pkl')
```

```python
df
```

```python
X = df.drop(columns = ['emotion', 'sentiment', 'tweet_text', 'tweet_tokens'])
y = df['sentiment']
```

```python
X.shape
```

```python
vectorizer = TfidfVectorizer()
tf_idf_features =  vectorizer.fit_transform(X['string_tokens']).toarray()
# test_tf_idf_features  =  vectorizer.transform(X_test['string_tokens']).toarray()

# Converting above list to DataFrame
tf_idf = pd.DataFrame(tf_idf_features)

# train_Y = df['sentiment']
# test_Y = test['label']

#Listing all features
features = ['char_count', 'word_count', 'sent_count',
       'capital_char_count', 'capital_word_count', 'quoted_word_count',
       'stopword_count', 'unique_word_count', 'htag_count', 'mention_count',
       'avg_wordlength', 'avg_sentlength', 'unique_vs_words',
       'stopwords_vs_words']

# Finally merging all features with above TF-IDF. 
X_train = pd.DataFrame.join(tf_idf, df[features])
# test  = pd.merge(test_tf_idf,X_test[features],left_index=True, right_index=True)
```

```python
len(X_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y, stratify= y, random_state=305)
```

# Models

```python
rfc = RandomForestClassifier()

rfc.fit(X_train, y_train)

rfc_y_preds = rfc.predict(X_test)
```
